```python
df2['movie_title'] = df2['original_title'].str.lower()
```

```python
# Parse the stringified features into their corresponding python objects

import ast
df2['genres'] = df2['genres'].map(lambda x: ast.literal_eval(x))
df2['cast'] = df2['cast'].map(lambda x: ast.literal_eval(x))
df2['crew'] = df2['crew'].map(lambda x: ast.literal_eval(x))
```

```python

def make_genresList(x):
    genres = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            genres.append(scifi)
        else:
            genres.append(i.get('name'))
    if genres == []:
        return np.NaN
    else:
        return (st.join(genres))
```

```python
df2['genres']=df2['genres'].map(lambda x:make_genresList(x))
```

```python
df2['genres']
```

```python

def get_actor1(x):
    casts = []
    for i in x:
       casts.append(i.get('name'))
        
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
       
        
```

```python
df2['actor_1_name'] = df2['cast'].map(lambda x: get_actor1(x))
```

```python
def get_actor2(x):
    casts=[]
    for i in x:
        casts.append(i.get('name'))
    if casts ==[] or len(casts)<=1:
        return np.NaN
    else:
        return casts[1]
```

```python
df2['actor_2_name'] = df2['cast'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    casts=[]
    for i in x:
        casts.append(i.get('name'))
    if casts ==[] or len(casts)<=2:
        return np.NaN
    else:
        return casts[2]
```

```python
df2['actor_3_name'] = df2['cast'].map(lambda x: get_actor3(x))
```

```python
def get_directors(x):
    directors=[]
    st=" "
    for i in x:
        if i.get('job')=='Director':
             directors.append(i.get('name'))
    if directors==[]:
        return np.NaN
    else:
        return st.join(directors)
    
       
```

```python
df2['directors']=df2['crew'].map(lambda x: get_directors(x))
```

```python
movie = df2.loc[:,['directors','actor_1_name','actor_2_name','actor_3_name','genres','title']]
```

```python
#drop null valued rows & columns
movie = movie.dropna(how='any')
```

```python
movie['combines_crew']=movie['actor_1_name']+' '+movie['actor_2_name']+' '+movie['actor_3_name']+' '+movie['directors']+' '+movie['genres']
```

```python
from tmdbv3api import Movie
tmdb_movie=Movie() #inititalise constructor
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
        genre_str = " " 
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres)
    else:
        np.NaN
    
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
def get_actor1(x):
    return (x.split("screenplay); ")[-1]).split(", ")[0]
```

```python
def get_actor2(x):
    if len(x.split("screenplay); ")[-1].split(", ")) < 2:
        return np.NaN
    else:
        return (x.split("screenplay); ")[-1]).split(", ")[1]
    
```

```python
def get_actor3(x):
    if len(x.split("screenplay); ")[-1].split(", ")) < 3:
        return np.NaN
    else:
        return (x.split("screenplay); ")[-1]).split(", ")[2]
```
